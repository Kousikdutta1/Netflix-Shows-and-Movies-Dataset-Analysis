#**Data Cleaning** -

```python
# Let's first visualize with the help of a Heatmap the null values present in our dataset

plt.figure(figsize = (10,5))
sns.heatmap(netflix_df.isnull());
plt.title('Null Values')
```

**We can see from above that the gaps present in the heatmap are the null values, we can clearly get an overview of the null values present.**

**it also can be said that  'director'  column has the maximum null values present.**

```python
# Let's fill the missing datas of Rating, Cast, Country, Director with 'Unavailable' 

netflix_df.fillna({'rating':'Unavailable', 'cast':'Unavailable', 'country':'Unavailable', 'director':"Unavailable"}, inplace = True)
```

```python
# Let's check for the total null values present in dataset according to columns 

netflix_df.isna().sum()
```

**It can be seen from above that Missing values were replaced by Unavailable.**

```python
# Let's find out where are the date_added is null

Date_Added_null = netflix_df[netflix_df.date_added.isnull()]
```

```python
print (Date_Added_null.shape)
```

**So, from above we can see that 10 rows are there where null values in date_added column is present.**

```python
# Let's fill the missing values of date_added with the most recent dates

date_added_null_values_rectification = netflix_df['date_added'].max()
print(date_added_null_values_rectification)
```

```python
# let's use fillna method to fillup the null values present in date_added column

netflix_df.fillna({'date_added':date_added_null_values_rectification}, inplace = True)
```

```python
# Let's check for the null values present in the dataset 

netflix_df.isna().sum()
```

**From above we can see that , successfully we have replaced all the null values except duration column which has only 3 null values present with it.**

```python
# Let's check the null values present on duration column

netflix_df[netflix_df['duration'].isna()]
```

**As we can see from the above data that null values are present in duration column, also we can encounter that duration is given in the rating column mistakenly, so we will be shifting the data of rating column to duration.**

```python
# From above we have seen that whatever duration values which are null are directed by Louis C.K.
# Let's check how many movies are directed by Louis C.K.

netflix_df[netflix_df['director'] == 'Louis C.K.']
```

**From above we can see that movies directed by Louis C.K. present in the dataset are only 3**

```python
# Let's shift the values of rating column to duration column for the movies directed by Louis C.K.

netflix_df.loc[netflix_df['director']=='Louis C.K.','duration'] = netflix_df['rating'] 
```

```python
# Now, Let's check for the dataset for the duration column


netflix_df[netflix_df['director']=='Louis C.K.']
```

**From above it can be seen that duration column has been replaced with the values of rating**

```python
# Now we have to change the ratings of Louis C.K. directed movies to Unavailable, as we have shifted its dataset to duration column

netflix_df.loc[netflix_df['director'] == 'Louis C.K.','rating'] = 'Unavailable'
```

```python
# Let's check for the Louis C.K directed movies 

netflix_df[netflix_df['director']=='Louis C.K.']
```

**From above we can conclude that, we have successfully shifted and changed the data of rating and duration columns**

```python
# Let's check for the null values if any present in our dataset

netflix_df.isna().sum()
```

**From above it is clear that we have cleaned the missing data present in dataset.**

```python
# Let's visualize with the help of a Heatmap the null values present in our dataset

plt.figure(figsize = (10,5))
sns.heatmap(netflix_df.isnull());
plt.title('Null Values')
```

#**4. Let's see the content added over the years on NETFLIX.**

```python
#Let's make a copy of the netflix dataset

netflix_df1 = netflix_df.copy()
```

```python
# Let's visualize the top 5 rows of the new netlfix dataset

netflix_df1.head()
```

```python
# Let's extract year from the date_added column

netflix_df1['year'] = pd.DatetimeIndex(netflix_df1['date_added']).year
```

```python
# Let's see the unique years available in the dataset

netflix_df1['year'].unique()
```

```python
# Let's create a new dataset with the Content type of Movie

new_dataset = netflix_df1[netflix_df1['type']=='Movie']
```

```python
new_dataset.head()
```

```python
# Let's group the distribution of Movie counts with respect to years

movies_dist_year = new_dataset.groupby(['year'])[['type']].count().reset_index()
```

```python
# Let's change the column names of the dataset

movies_dist_year.rename(columns = {'type' : 'movie_counts'}, inplace = True)
```

```python
# Let's see the movies distribution with year

movies_dist_year
```

```python
# Let's create a new dataset with the Content type of TV show

new_dataset2 = netflix_df1[netflix_df1['type']=='TV Show']
```

```python
# Let's see the top 5 rows

new_dataset2.head()
```

```python
# Let's group the distribution of TV Show Counts with respect to years

tvshow_dist_year = new_dataset2.groupby(['year'])[['type']].count().reset_index()
```

```python
# Let's change the column names of the dataset

tvshow_dist_year.rename(columns = {'type' : 'tvshow_counts'}, inplace = True)
```

```python
# Let's see the TV show distribution with year

tvshow_dist_year
```

```python
# Let's do outer merge both tvshow_dist_year, movies_dist_year dataset

merged_dataset = pd.merge (movies_dist_year,tvshow_dist_year, how = 'outer')
```

```python
# let's visualize the merged dataset

merged_dataset
```

```python
# Let's visualize with a barplot the distribution of Movies and Tv shows with respect to years 

merged_dataset.plot(x="year", y=['movie_counts','tvshow_counts'], kind="bar", figsize=(15, 8));
 
```

```python
# Let's visualize with a lineplot the distribution of Movies and Tv shows with respect to years 

merged_dataset.plot(x="year", y=['movie_counts','tvshow_counts'], kind="line", figsize=(15, 8));
```

**The growth in number of movies on netflix is much higher than that od TV shows. The growth in content started from 2013. Netflix kept on adding different movies and tv shows on its platform over the years.**

#**9. Netflix Film Duration.**

```python
# Let's extract the duration in integer format.

netflix_df['duration_in_numbers'] = netflix_df['duration'].str.extract('(\d+)')
```

```python
# Let's see the dataset after adding a new duration column.

netflix_df.head()
```

```python
# Let's create 2 separate dataset for Movies and TV Show


netflix_movies = netflix_df[netflix_df['type'] == 'Movie']
```

```python
netflix_tv_show = netflix_df[netflix_df['type'] == 'TV Show']
```

```python
netflix_movies['duration_in_numbers'].reset_index()
```

```python
# Let's visualize both TV Show and Movies duration plots
# Let's First define axes
fig,axes = plt.subplots(1,2,figsize=(18,6))

# Distribution curve for Movies duration
sns.distplot(netflix_movies['duration_in_numbers'], color = 'red', ax = axes[0] );
plt.xlabel('Duration of Movies')
plt.ylabel('Density')
plt.title('Distribution curve for Duration of Movies')

# Count plot for Number of seasons of TV Shows 
sns.countplot(x = netflix_tv_show['duration_in_numbers'], order = netflix_tv_show['duration_in_numbers'].value_counts().index[:10], color = 'red', ax = axes[1] );
plt.xlabel('Seasons')
plt.ylabel('Counts')
plt.title('Number of Seasons for TV Show')
plt.show();
```

**Now we will look into the duration of Netflix films. Since movies are measured in time and shows are measured by seasons, we need to split the dataset between movies and TV shows. On the left, we can see that the duration for Netflix movies closely resembles a normal distribution with the average viewing time spanning about 90 minutes which seems to make sense. Netflix TV shows on the other hand seems to be heavily skewed to the right where the majority of shows only have 1 season.**
